# A brief look at the statistics of emotions in lyrics

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
sns.set(style="darkgrid")

data_df = pd.read_csv('Cleaned_data.csv')
```

```python
# Helper Functions
def dec_to_percent(x):
    return x * 100
def milli_to_sec(x):
    return (x / 1000)/60
def lyrics_length(x):
    return len(x.split())
```

```python
# Feature engineering for columns
# To get a bit more interesting, let;s make a count of words to see if that has a correlation with emotion
data_df['word_count'] = data_df['cleaned_lyrics'].apply(lyrics_length)
data_df
```

## Anger emotion column

```python
data_df['anger']= data_df['anger'].apply(dec_to_percent)
data_df['anger'].describe()
```

```python
sns.relplot(x="anger", y="duration", data=data_df)
```

```python
sns.relplot(x="anger", y="word_count", data=data_df)
```

### Notes:
The 'angriest" songs seems to be highest in songs with a duration of about 3 and a half minutes, but that's where most songs lie in terms of duration anyway.

# Text data visualizations what are the top songs by emotion talking about?

## The angriest song

```python
data_df.sort_values(by=['anger'],ascending=False).head(1)['title']
```

```python
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(list(data_df.sort_values(by=['anger'],ascending=False).head(1)['cleaned_lyrics'])[0]) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
